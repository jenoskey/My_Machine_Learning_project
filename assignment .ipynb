##Define a dictionary  called students

```python
students={}
```

#Create class call (student)

```python
class student:
  def student_info(self,id,student_name,student_age,gander,address,city,data={},**student_grade):
    data = student_grade
    students[id]= {"student name" : student_name ,
                   "student age" :student_age ,
                   "gander" : gander ,"address" : address ,"city":city ,
                   "student_grade" : data }
#____________________________________________________________________________    
  
  def student_evaluation(self):
    x_id = str(input("type 'all' for all students grade or just specific 'id' "))
    if x_id == "all":
      print("the grade for all student is :")
      for i in range(1,4):
        x = int(students[str(i)]["student_grade"]["Math"])+int(students[str(i)]["student_grade"]["Science"])+int(students[str(i)]["student_grade"]["English"])
        print(f'the grade of student {students[str(i)]["student name"]} is ',x/3)
    else:
      x = int(students[x_id]["student_grade"]["Math"]) + int(students[x_id]["student_grade"]["Science"])  + int(students[x_id]["student_grade"]["English"])
      print(f'the grade of student {students[x_id]["student name"]} is ',x/3)
#____________________________________________________________________________

  def print_info(self):
    x_id = str(input("type 'all' for all students information or just specific 'id' "))
    if x_id == "all":
      for key,value in students.items():
        print(key,value)
    else:
      print(f'the information of student is {x_id}', students[x_id])
```

#Creat object



```python
Students = student()
```

#Define three students using an input method with a (for loop)

```python
for i in range(1,4):
  id = str(i)
  student_name= input(f"The 'id' for this student is : {i} \nEnter the student name : ")
  student_age = input("enter the age : ")
  gander = input("enter the gender : ")
  address = input("enter the address : ")
  city = input("enter the city : ")
  Math = input("enter the garde of Math  : ")
  Science = input("enter the grade of Science : ")
  English = input("enter the grade of English : ")
  Students.student_info(id,student_name,student_age,gander,address,city,Math = Math,Science = Science,English= English)
  
```
