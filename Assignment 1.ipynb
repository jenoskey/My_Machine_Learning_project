## 5 - Create a method inside the Student class call (student_evaluation), so it returns the mean of the student_grade.


```python
class Student:
   def __init__(self,student_name,student_age,gender ,address,city,student_grade):
     
    self.student_name=student_name
    self.student_age=student_age
    self.gender=gender
    self.address=address
    self.city=city
    self.student_grade = student_grade


   def student_info(self):
     
     return {'student_name':self.student_name, 'student_age':self.student_age, 
             'gender':self.gender, 'address':self.address, 'city':self.city,
             'student_grade':self.student_grade}
  

   def student_evaluation(self):
     

     """ Here we use both python's built in functions (sum, len) to help us in finding the mean
        sum function take the list of numbers as an input value and return their sum as an output
        len function take a list and return the number of item of the list as an output"""

     the_mean_value = sum(self.student_grade.values())/len(self.student_grade.values())

     return the_mean_value



   def __repr__(self): 
     
     """ this method will print the student informations once it is used with print function 
     for instance: >>> student = Student(arg)
                   >>> print(student) -> this will print out the returned string"""

     info = ''

     dict_info = {'student_name':self.student_name, 'student_age':self.student_age,
                  'gender':self.gender, 'address':self.address, 'city':self.city}

     for keys, value in dict_info.items(): 
       info += keys + ' : ' + str(value) +'\n'

     info += '-'*100 + '\n'

     for grade_keys, grade_value in self.student_grade.items():
       info +=  grade_keys + ' : ' + str(grade_value) +'\n'

     return info
```
